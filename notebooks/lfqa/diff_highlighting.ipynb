```python
import difflib

def show_matching_blocks(a, b):
    s = difflib.SequenceMatcher(None, a, b)
    m = s.get_matching_blocks()
    seqs = [a, b]

    new_seqs = []
    for select, seq in enumerate(seqs):
        i, n = 0, 0
        new_seq = ''
        while i < len(seq):
            if i == m[n][select]:
                new_seq += '{' + seq[m[n][select]:m[n][select] + m[n].size] + '}'
                i += m[n].size
                n += 1
            elif i < m[n][select]:
                new_seq += seq[i:m[n][select]]
                i = m[n][select]
        new_seqs.append(new_seq)
    for seq, n in zip(seqs, new_seqs):
        print('{} --> {}'.format(seq, n))
    print('')


show_matching_blocks(data.iloc[0]['gold_answer'], data.iloc[0]['fine_tuning'])
show_matching_blocks(data.iloc[0]['fine_tuning'], data.iloc[0]['gold_answer'])
show_matching_blocks(data.iloc[0]['text'], data.iloc[0]['fine_tuning'])
show_matching_blocks(data.iloc[0]['fine_tuning'], data.iloc[0]['text'])
```
